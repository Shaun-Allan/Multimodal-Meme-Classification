```python
from googletrans import Translator

translator = Translator()
```

```python
train_data["transcriptions"] = train_data["transcriptions"].apply(lambda x: translator.translate(x, src="ml", dest="en").text)

train_data
```

```python
train_data["labels"].loc[train_data.labels == "Misogyny"] = "misogyny"
train_data["labels"].loc[train_data.labels == "Not-misogyny"] = "not-misogyny"
train_data["labels"].loc[train_data.labels == "Not-Misogyny"] = "not-misogyny"
```

```python
import re
def remove_emojis(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
train_data["transcriptions"] = train_data["transcriptions"].apply(remove_emojis)
train_data
```

```python
import string
punctuation = string.punctuation.replace(",","")
def remove_punctuation(text):
    clean="".join([i for i in text if i not in punctuation])
    return clean
```

```python
train_data["transcriptions"] = train_data["transcriptions"].apply(remove_punctuation)
train_data
```

```python

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = stopwords.words('english')

# Define a function to remove stop words from a sentence
def remove_stop_words(sentence):
  # Split the sentence into individual words
  words = sentence.split()

  # Use a list comprehension to remove stop words
  filtered_words = [word for word in words if word not in stop_words]

  # Join the filtered words back into a sentence
  return ' '.join(filtered_words)
```

```python
train_data["transcriptions"] = train_data["transcriptions"].apply(remove_stop_words)
train_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train_text = vectorizer.fit_transform(train_data["transcriptions"])
```

```python
train_data["labels"].loc[train_data.labels == "Not Misogyny"] = 0
train_data["labels"].loc[train_data.labels == "Misogyny"] = 1
```

```python
print(train_data.loc[train_data.labels == 0].count())
print(train_data.loc[train_data.labels == 1].count())
```

```python
train_data_minority = train_data.loc[train_data.labels == 1]
```

```python
from sklearn.utils import resample

train_data_minority_upsampled = resample(train_data_minority, replace=True, n_samples = 200, random_state=42)
train_data_minority_upsampled.count()
```

```python
train_data_minority_upsampled
```

```python
train_data = pd.concat([train_data, train_data_minority_upsampled])
train_data
```

```python
print(train_data.loc[train_data.labels == 0].count())
print(train_data.loc[train_data.labels == 1].count())
```

```python
y_train = train_data["labels"]
y_train = y_train.infer_objects()
y_train
```

```python
from sklearn.naive_bayes import MultinomialNB

text_model = MultinomialNB()
text_model.fit(X_train_text, y_train)
```

```python
y_pred_text_train = text_model.predict(X_train_text)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_train, y_pred_text_train)
classification_rep = classification_report(y_train, y_pred_text_train)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)
```

```python
from sklearn.linear_model import LogisticRegression

text_model_lr = LogisticRegression()
text_model_lr.fit(X_train_text, y_train)
y_pred_text_train = text_model_lr.predict(X_train_text)
accuracy = accuracy_score(y_train, y_pred_text_train)
classification_rep = classification_report(y_train, y_pred_text_train)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)
```

```python
from sklearn.svm import SVC

text_model_svm = SVC()
text_model_svm.fit(X_train_text, y_train)
y_pred_text_train = text_model_svm.predict(X_train_text)
accuracy = accuracy_score(y_train, y_pred_text_train)
classification_rep = classification_report(y_train, y_pred_text_train)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)
```
